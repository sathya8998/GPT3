```python
from google.colab import files
from bs4 import BeautifulSoup

def format_html(html_content):
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all paragraphs and concatenate their text
    paragraphs = soup.find_all('p')
    formatted_text = '\n'.join([paragraph.get_text() for paragraph in paragraphs])

    return formatted_text

# Upload the text file
uploaded = files.upload()

# Get the file name
file_name = list(uploaded.keys())[0]

# Read the content of the file
with open(file_name, 'r') as file:
    html_content = file.read()

# Call the function to format HTML and extract text
formatted_text = format_html(html_content)

# Print or use the formatted text as needed
print(formatted_text)

# Save the formatted text to a new file
output_file_name = 'formatted_text.txt'
with open(output_file_name, 'w') as output_file:
    output_file.write(formatted_text)

# Download the formatted text file
files.download(output_file_name)
```
